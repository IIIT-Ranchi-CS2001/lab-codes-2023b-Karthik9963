# using numpy to calculate mean aoi,max pm2.5,min pm10 values for each city

```python
data = df[['City', 'AQI', 'PM2.5', 'PM10']].dropna().to_numpy()


cities = np.unique(data[:, 0])


result = {}
for city in cities:
    city_data = data[data[:, 0] == city][:, 1:].astype(float) 
    result[city] = {
        'Me AQI': np.mean(city_data[:, 0]),
        'Me PM2.5': np.max(city_data[:, 1]),
        'Me PM10': np.min(city_data[:, 2])
    }


for city, stats in result.items():
    print(f"City: {city}")
    print(f"  Mean AQI: {stats['Me AQI']}")
    print(f"  Max PM2.5: {stats['Me PM2.5']}")
    print(f"  Min PM10: {stats['Me PM10']}")
```
